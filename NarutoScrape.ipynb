```python
import pandas as pd
import twint
import nest_asyncio
nest_asyncio.apply()
```

```python

c=twint.Config()
c.Search = "Naruto"
c.TwitterSearch
c.Limit= 100000
c.Store_csv=True
c.Pandas = True
c.Output = "Naruto.csv"
c.Lang = "en"
twint.run.Search(c)
```
